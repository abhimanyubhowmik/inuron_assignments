### 3 . Create a table called as car with a column name given in dataset description

```python
cursor.execute('create table cardataset.car(buying VARCHAR(20),maint VARCHAR(20),door VARCHAR(20),persons VARCHAR(20),lug_boot VARCHAR(20),safty VARCHAR(20),rating VARCHAR(20))')
```

### 4 . Dump all the data into car table 

```python
with open('car.data','r') as car_data:
    for car in car_data:
        car_list = car.split(',')
        c_str = ''
        for c in range(len(car_list) - 1):
            c_str += '"{val}"'.format(val = car_list[c]) + ','
        c_str +=  '"{val}"'.format(val = car_list[-1])
        cursor.execute('insert into cardataset.car values ({car})'.format(car = c_str))
db.commit()
```
